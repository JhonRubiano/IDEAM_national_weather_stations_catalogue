```python
import requests
url = 'https://www.datos.gov.co/resource/hp9r-jxuu.json?$query=SELECT%0A%20%20%60codigo%60%2C%0A%20%20%60nombre%60%2C%0A%20%20%60categoria%60%2C%0A%20%20%60tecnologia%60%2C%0A%20%20%60estado%60%2C%0A%20%20%60departamento%60%2C%0A%20%20%60municipio%60%2C%0A%20%20%60ubicaci_n%60%2C%0A%20%20%60altitud%60%2C%0A%20%20%60fecha_instalacion%60%2C%0A%20%20%60fecha_suspension%60%2C%0A%20%20%60area_operativa%60%2C%0A%20%20%60corriente%60%2C%0A%20%20%60area_hidrografica%60%2C%0A%20%20%60zona_hidrografica%60%2C%0A%20%20%60subzona_hidrografica%60%2C%0A%20%20%60entidad%60'
json = requests.get(url).json()
```

Convierto el archivo json en un data framer de python

```python
import pandas as pn
data = pn.DataFrame(json)
```

Acá se muestra los datos que se han llamado, la intención es corregir la columna nombre, la ubiación.

```python
data.head(5)
```

Limpieza columna nombre

```python
listaNombres = []
for i in  range(0,len(data['nombre'])):
    if data['nombre'][i].find(' -') != -1:
        nombre = data['nombre'][i].split(' -')
        listaNombres.append(nombre[0])
    else:
        nombre = data['nombre'][i].split(' [')
        listaNombres.append(nombre[0])
data['nombre'] = listaNombres
data.head(10)
```

Correción de la columna ubicación

```python
data.ubicaci_n
latitude = []
longitude = []
for i in data.ubicaci_n:
    latitude.append(float(i['latitude']))
    longitude.append(float(i['longitude']))
```

Creo dos nuevas columnas con los nombres de latitude y longited, puede ser utiles para un grafico de map

```python
data['latitude'] = latitude
data['longitude'] = longitude
```

Elimino el la columna ubicación

```python
data = data.drop(columns='ubicaci_n')
```

Datos finales

```python
dataCsv = data.to_csv('Ambiente Y Desarrollo Sostenible', index=False) 
data.head(10)
```
